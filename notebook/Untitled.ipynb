```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def ode(x, y, t, a, b, c, d):
        dydt = [a*x - b*x*y, c*x*y - d*y]
        return dydt
        

def calcODE(args, y0, dy0, ts = 10, nt = 101):
    y0 = 100
    x0 = 100
    t = np.linspace(0, ts, nt)
    sol = odeint(ode, y0, x0, t, args)
    return sol
   
def drawPhasePortrait(args, deltaX = 1, deltaDX = 1, startX = 100,  stopX = 5, startDX = 0, stopDX = 5, ts = 10, nt = 101):
    for y0 in range(startX, stopX, deltaX):
            for dy0 in range(startDX, stopDX, deltaDX):
                sol = calcODE(args, y0, dy0, ts, nt)
                plt.plot(sol[:, 0], sol[:, 1], 'b')
    plt.xlabel('x')
    plt.ylabel('dx/dt')
    plt.grid()
    plt.show()
```

```python
a = 4
b = 2.5
c = 2
d = 1
args=(a, b, c, d)
```

```python
drawPhasePortrait(args)
```

```python
drawPhasePortrait(args, 1, 1, -10, 10, -5, 5, ts = 30, nt = 301)
```
