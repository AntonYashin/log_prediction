```python
import numpy as np
import matplotlib.pyplot as plt

r_dt=np.arange(1, 3.01, 0.01) 

for i in range(len(r_dt)):
    X0 = 0.1
    x_Axis = np.array([])
    X = np.array([])        
    for j in range(4200):   
        x_Axis = np.append(x_Axis, r_dt[i])
        X = np.append(X, X0 + r_dt[i]*(1 - X0)*X0)
        X0 = X[j]
        #print("Length x-axis = ", len(x_Axis), "Length X = ", len(X[1000:]))
    plt.scatter(x_Axis[4000:],X[4000:],c='green')
    
plt.xlabel(r'$r_dt$') 
plt.ylabel(r'$f(r_dt)$') 
plt.title(r'$f_1(xr_dt)=\sin(r_dt),\ f_2(r_dt)=\cos(r_dt),\ f_3(r_dt)=-r_dt$')
plt.grid(True)
plt.show()
```
